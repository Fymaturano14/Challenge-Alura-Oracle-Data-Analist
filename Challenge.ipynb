```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Combinar los datos

### Subtask:
Unir los cuatro dataframes en uno solo para facilitar el análisis conjunto.


**Reasoning**:
The subtask requires combining the four dataframes into a single one. I will create a list of the dataframes and then concatenate them using pandas. Finally, I will display the head of the combined dataframe to confirm the operation.



```python
dataframes = [tienda, tienda2, tienda3, tienda4]
df_combined = pd.concat(dataframes, ignore_index=True)
display(df_combined.head())
```

## Convertir la columna 'fecha de compra' a datetime

### Subtask:
Asegurarse de que la columna 'Fecha de Compra' esté en el formato datetime para realizar análisis temporales.


**Reasoning**:
Convert the 'Fecha de Compra' column to datetime objects for temporal analysis.



```python
df_combined['Fecha de Compra'] = pd.to_datetime(df_combined['Fecha de Compra'], format='%d/%m/%Y')
display(df_combined.head())
```

## Crear una columna 'total de la compra'

### Subtask:
Calcular el total de cada compra sumando el 'Precio' y el 'Costo de envío'.


**Reasoning**:
Create a new column 'Total de la Compra' by adding 'Precio' and 'Costo de envío' and then display the head of the dataframe to verify the result.



```python
df_combined['Total de la Compra'] = df_combined['Precio'] + df_combined['Costo de envío']
display(df_combined.head())
```

## Análisis de ventas a lo largo del tiempo

### Subtask:
Analizar cómo varían las ventas totales, el precio promedio y el costo de envío promedio a lo largo del tiempo (por ejemplo, por mes o año).


**Reasoning**:
Extract the year and month from the 'Fecha de Compra' column and calculate the sum of 'Total de la Compra', average of 'Precio', and average of 'Costo de envío' grouped by year and month.



```python
df_combined['Año'] = df_combined['Fecha de Compra'].dt.year
df_combined['Mes'] = df_combined['Fecha de Compra'].dt.month
temporal_analysis = df_combined.groupby(['Año', 'Mes']).agg(
    Total_Ventas=('Total de la Compra', 'sum'),
    Precio_Promedio=('Precio', 'mean'),
    Costo_Envio_Promedio=('Costo de envío', 'mean')
).reset_index()
display(temporal_analysis.head())
```

## Análisis por categoría de producto

### Subtask:
Analizar las ventas totales, el precio promedio y el costo de envío promedio por categoría de producto.


**Reasoning**:
Group the combined dataframe by product category and calculate the sum of total sales, the average price, and the average shipping cost for each category.



```python
category_analysis = df_combined.groupby('Categoría del Producto').agg(
    Total_Ventas_Categoria=('Total de la Compra', 'sum'),
    Precio_Promedio_Categoria=('Precio', 'mean'),
    Costo_Envio_Promedio_Categoria=('Costo de envío', 'mean')
).reset_index()
display(category_analysis.head())
```

## Análisis por lugar de compra

### Subtask:
Analizar las ventas totales, el precio promedio y el costo de envío promedio por lugar de compra.


**Reasoning**:
Group the dataframe by location and calculate the required aggregations.



```python
location_analysis = df_combined.groupby('Lugar de Compra').agg(
    Total_Ventas=('Total de la Compra', 'sum'),
    Precio_Promedio=('Precio', 'mean'),
    Costo_Envio_Promedio=('Costo de envío', 'mean')
).reset_index()
display(location_analysis.head())
```

## Análisis por método de pago y cantidad de cuotas

### Subtask:
Analizar cómo influyen el método de pago y la cantidad de cuotas en las ventas.


**Reasoning**:
Group the data by payment method and installment count and calculate the sum of total purchase, then reset the index and display the head of the resulting dataframe.



```python
payment_installment_analysis = df_combined.groupby(['Método de pago', 'Cantidad de cuotas']).agg(
    Total_Ventas=('Total de la Compra', 'sum')
).reset_index()
display(payment_installment_analysis.head())
```

## Análisis de calificaciones

### Subtask:
Analizar la distribución de las calificaciones y cómo se relacionan con otros factores (por ejemplo, categoría de producto, vendedor).


**Reasoning**:
Calculate the distribution of ratings and the average rating by category and seller, then store the results in dataframes and display their heads.



```python
rating_distribution = df_combined['Calificación'].value_counts().sort_index()
category_rating_analysis = df_combined.groupby('Categoría del Producto')['Calificación'].mean().reset_index()
seller_rating_analysis = df_combined.groupby('Vendedor')['Calificación'].mean().reset_index()

display(rating_distribution)
display(category_rating_analysis.head())
display(seller_rating_analysis.head())
```

## Análisis por vendedor

### Subtask:
Analizar las ventas por vendedor.


**Reasoning**:
Group the dataframe by seller and calculate the sum of total sales.



```python
seller_sales_analysis = df_combined.groupby('Vendedor')['Total de la Compra'].sum().reset_index()
display(seller_sales_analysis.head())
```

```python
product_sales = df_combined.groupby('Producto')['Total de la Compra'].sum().reset_index()

most_sold_products = product_sales.sort_values(by='Total de la Compra', ascending=False)
least_sold_products = product_sales.sort_values(by='Total de la Compra', ascending=True)

print("Productos más vendidos:")
display(most_sold_products.head())

print("\nProductos menos vendidos:")
display(least_sold_products.head())
```
